```python
from transformers import pipeline
```

```python
generator = pipeline('text-generation', model='EleutherAI/gpt-neo-1.3B')
```

```python
input_prompt = input("Enter prompt to generate text:\n")
```

```python
max_length = int(input("Enter max legnth constraint to generate the text:\n"))
```

```python
result = generator(input_prompt, max_length=max_length, do_sample=True, temperature=0.9)
```
